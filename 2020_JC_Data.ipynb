```python
import pandas as pd
```

```python
#load in Jersey City csvs from https://s3.amazonaws.com/tripdata/index.html
JC_data_Jan = pd.read_csv("JC-202001-citibike-tripdata.csv")
JC_data_Feb =pd.read_csv("JC-202002-citibike-tripdata.csv")
JC_data_Mar =pd.read_csv("JC-202003-citibike-tripdata.csv")
JC_data_Apr =pd.read_csv("JC-202004-citibike-tripdata.csv")
JC_data_May =pd.read_csv("JC-202005-citibike-tripdata.csv")
JC_data_Jun =pd.read_csv("JC-202006-citibike-tripdata.csv")
JC_data_July =pd.read_csv("JC-202007-citibike-tripdata.csv")
JC_data_Aug =pd.read_csv("JC-202008-citibike-tripdata.csv")
```

```python
#combine csvs
total_2020_JC = pd.concat([JC_data_Jan, JC_data_Feb, JC_data_Mar, JC_data_Apr,JC_data_May, JC_data_Jun, JC_data_July,JC_data_Aug])
total_2020_JC
```

```python
#output to csv for load into Tableau
total_2020_JC.to_csv("Combined_JC_2020.csv")
```
