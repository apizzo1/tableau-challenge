```python
import pandas as pd
```

```python
#load Jersey City csvs from https://s3.amazonaws.com/tripdata/index.html
JC_data_Jan = pd.read_csv("JC-201901-citibike-tripdata.csv")
JC_data_Feb = pd.read_csv("JC-201902-citibike-tripdata.csv")
JC_data_Mar = pd.read_csv("JC-201903-citibike-tripdata.csv")
JC_data_Apr = pd.read_csv("JC-201904-citibike-tripdata.csv")
JC_data_May = pd.read_csv("JC-201905-citibike-tripdata.csv")
JC_data_Jun = pd.read_csv("JC-201906-citibike-tripdata.csv")
JC_data_July = pd.read_csv("JC-201907-citibike-tripdata.csv")
JC_data_Aug = pd.read_csv("JC-201908-citibike-tripdata.csv")
```

```python
#combine the csvs
total_2019_JC = pd.concat([JC_data_Jan, JC_data_Feb, JC_data_Mar, JC_data_Apr,JC_data_May, JC_data_Jun, JC_data_July,JC_data_Aug])
total_2019_JC
```

```python
# output to a csv to load into Tableau
total_2019_JC.to_csv("Combined_JC_2019.csv")
```
