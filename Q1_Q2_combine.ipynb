```python
import pandas as pd
```

```python
#load q1 csvs
bike_data_2014 = pd.read_csv("Q1_2014.csv")
bike_data_2015 = pd.read_csv("Q1_2015.csv")
bike_data_2016 = pd.read_csv("Q1_2016.csv")
bike_data_2017 = pd.read_csv("Q1_2017.csv")
bike_data_2018 = pd.read_csv("Q1_2018.csv")
bike_data_2019 = pd.read_csv("Q1_2019.csv")
```

```python
bike_data_2017
```

```python
bike_data_2014 = bike_data_2014.rename(columns={"Total Annual Members":"Total Annual Members (All Time)"})
bike_data_2014
```

```python
bike_data_2014_q2 = pd.read_csv("Q2_2014.csv")
bike_data_2015_q2 = pd.read_csv("Q2_2015.csv")
bike_data_2016_q2 = pd.read_csv("Q2_2016.csv")
bike_data_2017_q2 = pd.read_csv("Q2_2017.csv")
bike_data_2018_q2 = pd.read_csv("Q2_2018.csv")
bike_data_2019_q2 = pd.read_csv("Q2_2019.csv")
```

```python
bike_data_2014_q2
```

```python
bike_data_2014_q2 = bike_data_2014_q2.rename(columns={"Total Annual Memberships Sold":"Total Annual Members (All Time)"})
bike_data_2014_q2
```

```python
bike_data_2015_q2 = bike_data_2015_q2.rename(columns={"Total Annual Memberships Sold":"Total Annual Members (All Time)"})
bike_data_2015_q2
```

```python
q1__q2_compare = pd.concat([bike_data_2014,bike_data_2015,bike_data_2016,bike_data_2017,bike_data_2018, bike_data_2019,\
                           bike_data_2014_q2,bike_data_2015_q2,bike_data_2016_q2,bike_data_2017_q2,bike_data_2018_q2,bike_data_2019_q2])
q1__q2_compare
```

```python
combined_q1_q2_compare=q1__q2_compare[["Date","Trips over the past 24-hours (midnight to 11:59pm)", "Miles traveled today (midnight to 11:59 pm)","24-Hour Passes Purchased (midnight to 11:59 pm)","Total Annual Members (All Time)","3-Day Passes Purchased (midnight to 11:59 pm)"]]
```
